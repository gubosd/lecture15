# OpenCV 이미지 특성

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
bob = cv2.imread('bob.jpg')
```

### 이미지는 숫자
- 각 픽셀은 0~255 의 정수임 (np.uint8 타입)

```python
bob.min(), bob.max()
```

```python
bob[:5,:5,0]
```

```python
bob.dtype
```

### 채널은 BGR 순서
- RGB 가 아닌 Blue-Green-Red 순서임
- cv2.imshow() 함수도 BGR 형태의 이미지를 입력받음

```python
bob2 = bob.copy()
bob2[:,:,2] = 0 # 특정 색상 제거
```

```python
cv2.imshow('image', bob2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- RGB 이미지로 전환

```python
bob2 = bob[:,:,[2,1,0]]
plt.imshow(bob2)
```

### 채널 분리

```python
blue = bob[:,:,0]
green = bob[:,:,1]
red = bob[:,:,2]
```

```python
plt.figure(figsize=[10,4])

plt.subplot(1,3,1)
plt.title('blue')
plt.imshow(blue, cmap='Blues')

plt.subplot(1,3,2)
plt.title('green')
plt.imshow(green, cmap='Greens')

plt.subplot(1,3,3)
plt.title('red')
plt.imshow(red, cmap='Reds')
```

### 이미지 자르기

```python
bob2 = bob[:200, 100:300]
cv2.imshow('image', bob2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 잘라서 붙여넣기

```python
bob2 = bob.copy()
bob2[100:300, 100:300] = bob[:200,:200]
cv2.imshow('image', bob2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 지우기

```python
bob2 = bob.copy()
bob2[50:150, 100:300] = 0
cv2.imshow('image', bob2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 밝기 반전

```python
cv2.imshow('image', 255-bob)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 대칭 변환

```python
cv2.imshow('image', bob[::-1])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('image', bob[:, ::-1])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 회전 변환

```python
bob2 = np.swapaxes(bob, 0, 1) # np.moveaxis(bob,0,1)
cv2.imshow('image', bob2[::-1])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
bob2 = np.swapaxes(bob, 0, 1)
cv2.imshow('image', bob2[:,::-1])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('image', bob[::-1,::-1])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 픽셀값 증가

```python
cv2.imshow('image', bob+128)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 이미지는 np.uint8 형태이므로 255에 1을 더하면 0이 된다

```python
bob2 = np.array(bob, dtype=np.uint16)
bob3 = np.minimum(bob2+128,255)
cv2.imshow('image', bob3.astype(np.uint8))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
