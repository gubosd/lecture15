## OpenCV
# 색공간 (color space)

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
bob = cv2.imread('bob.jpg')
```

### 채널 분리
- cv2.split()
- cv2.merge()

```python
R,G,B = cv2.split(bob)
```

```python
plt.figure(figsize=[10,5])

plt.subplot(1,3,1)
plt.title('Blue')
plt.imshow(B, cmap='gray')

plt.subplot(1,3,2)
plt.title('Green')
plt.imshow(G, cmap='gray')

plt.subplot(1,3,3)
plt.title('Red')
plt.imshow(R, cmap='gray')
```

```python
zeros = np.zeros(bob.shape[:2], dtype='uint8')
bob2 = cv2.merge([B, zeros, zeros])
plt.imshow(bob2[:,:,[2,1,0]])
```

```python
zeros = np.zeros(bob.shape[:2], dtype='uint8')
bob2 = cv2.merge([zeros, G, zeros])
plt.imshow(bob2[:,:,[2,1,0]])
```

```python
zeros = np.zeros(bob.shape[:2], dtype='uint8')
bob2 = cv2.merge([zeros, zeros, R])
plt.imshow(bob2[:,:,[2,1,0]])
```

### HSV 색공간
- 색상 (hue) : 0~179 (색상 종류)
- 채도 (saturation) : 0~255 (맑은 정도)
- 명도 (value) : 0~255 (밝은 정도)

```python
bob_hsv = cv2.cvtColor(bob, cv2.COLOR_BGR2HSV)
```

```python
bob_hsv.shape, bob_hsv.max(), bob_hsv.min()
```

```python
bob_hsv[:,:,0].max()
```

```python
plt.subplot(1,2,1)
plt.imshow(bob[:,:,[2,1,0]])

plt.subplot(1,2,2)
plt.imshow(bob_hsv) # 파란색이 많을 수록 밝은 영역이다
```
